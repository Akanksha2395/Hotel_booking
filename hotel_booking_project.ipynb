```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
import plotly.express as px
```

```python
df['hotel'].value_counts().plot(kind='pie',shadow=True,explode=(0,0.2),startangle = 90,autopct='%1.1f%%')
plt.title('the large no of used hotels')
plt.legend()
```

```python
plt.figure(figsize=(20,10))
df.arrival_date_year.value_counts(normalize=True).plot(kind='bar',color = ['orange','gray','green'])
plt.title('year_of_customers_arrive')
plt.xlabel('years')       
plt.ylabel('frequency')                                 
plt.show()
plt.legend()
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x='arrival_date_month', hue = 'hotel', data= df)
plt.title('Months of Arrival')
plt.show()
```

```python
top10 = df[df['country'].isin(["PRT","GBR","FRA","ESP","DEU","ITA","IRL",
                         "BEL","BRA","NLD"])]
```

```python
temp = df['country'].value_counts().reset_index().rename(columns={'index':'country','country':'count'})
plt.rcParams['figure.figsize'] = (60, 10)
guest_map = px.choropleth(temp,
                          locations=temp['country'],
                          color=np.log(temp['count']), 
                          hover_name=temp['country'], 
                          color_continuous_scale=px.colors.sequential.Plasma,
                          title="Home country of guests")

guest_map.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
df['market_segment'].value_counts().plot(kind='area',)
plt.title('booked hotel from market segment')
plt.xlabel('market segment')
plt.ylabel('total no of booking')
```

```python
sns.countplot(x='is_canceled',hue='hotel', data= df)
plt.title('Cancellation Situation')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
df['is_repeated_guest'].value_counts().plot(kind='barh',color=['red','yellow'])
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
df['deposit_type'].value_counts().plot(kind='hist')
plt.title('customer deposits')
plt.xlabel('deposit  type')
plt.ylabel('total no of money')
```

```python
fig, ax = plt.subplots(figsize=(22,15))
sns.heatmap(df.corr(),annot=True,ax=ax);
```
